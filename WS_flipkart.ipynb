In web scraping, a status code of 200 typically indicates that the HTTP request sent to the website was successful and the server has responded with the requested data.

The HTTP status code 200 is known as a "OK" response, indicating that the server has received the request and processed it successfully. It means that the web page or resource you were trying to access was found and returned by the server.

This is a desirable status code for web scraping because it means that the scraper was able to access and retrieve the data it needed from the website without any issues. Other HTTP status codes, such as 404 (Not Found), 403 (Forbidden), or 500 (Internal Server Error) may indicate that there was a problem with the request or that the server was unable to fulfill it.

In BeautifulSoup, the `soup.p` code is used to access the first `p` element in the parse tree created from the HTML content. 

The `p` element in HTML is used to define a paragraph of text. It is a block-level element that is typically used to group together related text content. For example, a paragraph might contain a single sentence or several sentences that discuss a particular topic or idea.

By accessing `soup.p`, the code is retrieving the first `p` element in the HTML document. If there are multiple `p` elements in the document, `soup.p` will only return the first one. If there are no `p` elements in the document, `soup.p` will return `None`.

Once the `p` element is retrieved, you can further process it to extract its contents or attributes using various methods and properties provided by BeautifulSoup. For example, you can use `soup.p.text` to get the text content of the `p` element or `soup.p['class']` to get the value of the `class` attribute of the `p` element.

```python
print(soup.p.text)
```

Once the `a` element is retrieved, you can further process it to extract its contents or attributes using various methods and properties provided by BeautifulSoup. For example, you can use `soup.a.text` to get the text content of the `a` element or `soup.a['href']` to get the value of the `href` attribute of the `a` element.

```python


from bs4 import BeautifulSoup
import requests 
```

```python
url = "https://www.flipkart.com/cameras/dslr-mirrorless/pr?sid=jek,p31,trv&otracker=categorytree"

response=requests.get(url)
htmlcontent=response.content
soup=BeautifulSoup(htmlcontent,'html.parser')
```

```python
titles = []
prices = []
images = []
```

find title from above div

```python
for d in soup.find_all('div', attrs={'class':'_2kHMtA'}): 
#     print(d) 
    title=d.find('div', attrs={'class':'_4rR01T'})
    print(title)
```

```python
for d in soup.find_all('div', attrs={'class':'_2kHMtA'}): 
#     print(d) 
    title=d.find('div', attrs={'class':'_4rR01T'})
    print(title.string)
```

for price


```python
for d in soup.find_all('div', attrs={'class':'_2kHMtA'}): 
#     print(d) 
    title=d.find('div', attrs={'class':'_4rR01T'})
    print(title.string)
    
    price=d.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
    print(price.string)
 
```

```python
for d in soup.find_all('div', attrs={'class':'_2kHMtA'}): 
#     print(d) 
    title=d.find('div', attrs={'class':'_4rR01T'})
    print(title.string)
    
    price=d.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
    print(price.string)
    
    image=d.find('img', attrs={'class':'_396cs4'})
    print(image.get('src'))
```

```python
for d in soup.find_all('div', attrs={'class':'_2kHMtA'}): 
#     print(d) 
    title=d.find('div', attrs={'class':'_4rR01T'})
#     print(title.string)
    
    price=d.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
#     print(price.string)
    
    image=d.find('img', attrs={'class':'_396cs4'})
#     print(image.get('src'))
    
    titles.append(title.string)
    prices.append(price.string)
    images.append(image.get('src'))
    
```
