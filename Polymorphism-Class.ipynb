### Polymorphism With Inheritance


```python
class Vehicle:

    def __init__(self, name, color, price):
        self.name = name
        self.color = color
        self.price = price

    def show(self):
        print('Details:', self.name, self.color, self.price)

    def max_speed(self):
        print('Vehicle max speed is 150')

    def change_gear(self):
        print('Vehicle change 6 gear')

        # inherit from vehicle class
class Car(Vehicle):
    def max_speed(self):
        print('Car max speed is 240')

    def change_gear(self):
        print('Car change 7 gear')
# Car Object
car = Car('Car x1', 'Red', 20000)
car.show()
# calls methods from Car class
car.max_speed()
car.change_gear()
# Vehicle Object
vehicle = Vehicle('Truck x1', 'white', 75000)
vehicle.show()
# calls method from a Vehicle class
vehicle.max_speed()
vehicle.change_gear()
```

####  Polymorphism In Class methods


```python
class Ferrari:
    def fuel_type(self):
        print("Petrol")
    def max_speed(self):
        print("Max speed 350")
class BMW:
    def fuel_type(self):
        print("Diesel")
    def max_speed(self):
        print("Max speed is 240")
ferrari = Ferrari()
bmw = BMW()
# iterate objects of same type
for car in (ferrari, bmw):
    # call methods without checking class of object
    car.fuel_type()
    car.max_speed()
```

### Polymorphism with Function and Objects


```python
class Ferrari:
    def fuel_type(self):
        print("Petrol")

    def max_speed(self):
        print("Max speed 350")

class BMW:
    def fuel_type(self):
        print("Diesel")

    def max_speed(self):
        print("Max speed is 240")

# normal function
def car_details(obj):
    obj.fuel_type()
    obj.max_speed()

ferrari = Ferrari()
bmw = BMW()

car_details(ferrari)
car_details(bmw)
```

#### Method Overloading


The process of calling the same method with different parameters is known as method overloading. Python does not support method overloading. Python considers only the latest defined method even if you overload the method. Python will raise a TypeError if you overload the method.



```python
def addition(a, b):
    c = a + b
    print(c)

def addition(a, b, c):
    d = a + b + c
    print(d)

# the below line shows an error
# addition(4, 5)
# This line will call the second product method
addition(3, 7, 5)
```

```python
for i in range(5): print(i, end=', ')
print()
for i in range(5, 10): print(i, end=', ')
print()
for i in range(2, 12, 2): print(i, end=', ')
```

```python
class Shape:
    # function with two default parameters
    def area(self, a, b=0):
        if b > 0:
            print('Area of Rectangle is:', a * b)
        else:
            print('Area of Square is:', a ** 2)

square = Shape()
square.area(5)

rectangle = Shape()
rectangle.area(5, 3)
```

In Python, method overloading is not directly supported like in some other programming languages such as Java or C++. However, we can achieve similar behavior using default parameter values and conditional statements, as demonstrated in the provided code.

The `Shape` class defines a method called `area` that calculates the area of a shape. It takes two parameters: `a` and `b`. The `a` parameter represents the base or side length of the shape, and the `b` parameter represents an optional second dimension (e.g., the height or width).

Inside the `area` method, there is a conditional statement that checks whether `b` is greater than zero. If it is, the method assumes that the shape is a rectangle and calculates the area by multiplying `a` and `b`. Otherwise, if `b` is zero or less, the method assumes that the shape is a square and calculates the area by squaring `a`.

Now, let's look at the method calls:

1. `square = Shape()`: This creates an instance of the `Shape` class named `square`.
2. `square.area(5)`: This calls the `area` method on the `square` object and passes the value 5 as the `a` parameter. Since no value is provided for `b`, the default value of 0 is used. The conditional statement in the `area` method checks that `b` is not greater than 0, so it assumes that the shape is a square and prints the area of the square, which is 5^2 = 25.

3. `rectangle = Shape()`: This creates another instance of the `Shape` class named `rectangle`.
4. `rectangle.area(5, 3)`: This calls the `area` method on the `rectangle` object and passes the values 5 and 3 as the `a` and `b` parameters, respectively. Since `b` is greater than 0, the conditional statement in the `area` method determines that the shape is a rectangle and calculates the area by multiplying `a` and `b`. It then prints the area of the rectangle, which is 5 * 3 = 15.

Although this approach does not strictly adhere to the concept of method overloading, it allows the `area` method to handle different scenarios based on the number of arguments passed. By using default parameter values and conditional statements, we can achieve similar functionality as method overloading in other languages.
