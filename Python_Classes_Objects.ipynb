#### Access Class Attributes Using Objects


We use the . notation to access the attributes of a class. For example,

```python
# define a class
class Bike:
    name = ""
    gear = 0

# create object of class
bike1 = Bike()

# access attributes and assign new values
bike1.gear = 11
bike1.name = "Mountain Bike"

print(f"Name: {bike1.name}, Gears: {bike1.gear} ")
```

```python
# create object of class
bike2 = Bike()

# access attributes and assign new values
bike2.gear = 6
bike2.name = "Royal Enfield"

print(f"Name: {bike2.name}, Gears: {bike2.gear} ")
```

#### PYTHON CONSTRUCTORS 

Constructors are generally used for instantiating an object. 

The task of constructors is to initialize(assign values) to the data members of the class when an object of the class is created. 

In Python the __init__() method is called as the constructor and **is always called when an object is created.**

#### Types of constructors : 

**default constructor**: The default constructor is a simple constructor which doesnt accept any arguments. Its definition has only one argument which is a reference to the instance being constructed.

**parameterized constructor**: constructor with parameters is known as parameterized constructor. The parameterized constructor takes its first argument as a reference to the instance being constructed known as self and the rest of the arguments are provided by the programmer.

```python
class college: 
    def __init__(self): 
        self.city ="Pune"
        self.university="sppu"
```

```python
c1=college()
```

```python
print(c1.city, c1.university)
```

```python
c2=college()
print(c2.city, c2.university)
```

```python
# Detault 
class Car:
     # default constructor
    def __init__(self):
        self.name = "Rolls Royce"
 
    # a method for printing data members
    def print_Car(self):
        print(self.name)
 
 
# creating object of the class
obj = Car()
 
# calling the instance method using the object obj
obj.print_Car()
```
