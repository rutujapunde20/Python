```python
def my_decorator(func):
    def wrapper():
        print("Before the function is called.")
        func()
        print("After the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
def my_decorator1(func):
    def wrapper():
        print("*********.")
        func()
        print("****************.")
    return wrapper

def my_decorator2(func):
    def wrapper():
        print("+++++++++")
        func()
        print("+++++++++++++++++++++")
    return wrapper

@my_decorator1
@my_decorator2
def say_hello():
    print("Hello!")

say_hello()
```
